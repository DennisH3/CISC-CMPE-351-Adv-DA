```python
import numpy as np
import pandas as pd
```

```python
pred = np.loadtxt('MLP_Classification_ypred_Financial.csv',delimiter=',')
test = np.loadtxt('MLP_Classification_ytest_Financial.csv', delimiter = ',')

columns = ["max1_5d","max1_10d","max1_15d","max1_20d","buy_5d","buy_10d","buy_15d","buy_20d"]

pred = pd.DataFrame(data = pred,columns = columns)
test = pd.DataFrame(data = test,columns = columns)
pred.head()
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

days = [5,10,15,20]

for i in days:
  p = pred["max1_{}d".format(i)]
  t = test["max1_{}d".format(i)]
  p2 = pred["buy_{}d".format(i)]
  t2 = test["buy_{}d".format(i)]
  print("{} Day Max Prediction Accuracy".format(i), accuracy_score(t,p))
  print("{} Day Max Prediction Precision".format(i), precision_score(t,p, average = 'micro'))
  print("{} Day Max Prediction Recall".format(i), recall_score(t,p, average = 'micro'))
  print("{} Day Max Prediction F1 Score".format(i), f1_score(t,p, average = 'micro'))
  print("-------------------------------------------------------------------")
  print("{} Day Buy Prediction Accuracy".format(i), accuracy_score(t2,p2))
  print("{} Day Buy Prediction Precision".format(i), precision_score(t2,p2, average = 'binary'))
  print("{} Day Buy Prediction Recall".format(i), recall_score(t2,p2, average = 'binary'))
  print("{} Day Buy Prediction F1 Score".format(i), f1_score(t2,p2, average = 'binary'))
  print("--------------------------------------------------------------------")
```

```python
# Empty lists
tol_5d = []

tol_10d = []
tol2_10d = []

tol_15d = []
tol2_15d = []
tol3_15d = []

tol_20d = []
tol2_20d = []
tol3_20d = []

for i in range(len(pred.index)):
  if (( (test.iloc[i]["max1_5d"] -1) <= pred.iloc[i]["max1_5d"] <= (test.iloc[i]["max1_5d"]+ 1)) == True):
    tol_5d.append(1)
  else:
    tol_5d.append(0)

  if (( (test.iloc[i]["max1_10d"] -1) <= pred.iloc[i]["max1_10d"] <= (test.iloc[i]["max1_10d"]+ 1)) == True):
    tol_10d.append(1)
  else:
    tol_10d.append(0)

  if (( (test.iloc[i]["max1_10d"] -2) <= pred.iloc[i]["max1_10d"] <= (test.iloc[i]["max1_10d"]+ 2)) == True):
    tol2_10d.append(1)
  else:
    tol2_10d.append(0)
  
  if (( (test.iloc[i]["max1_15d"] -1) <= pred.iloc[i]["max1_15d"] <= (test.iloc[i]["max1_15d"]+ 1)) == True):
    tol_15d.append(1)
  else:
    tol_15d.append(0)
  
  if (( (test.iloc[i]["max1_15d"] -2) <= pred.iloc[i]["max1_15d"] <= (test.iloc[i]["max1_15d"]+ 2)) == True):
    tol2_15d.append(1)
  else:
    tol2_15d.append(0)

  if (( (test.iloc[i]["max1_15d"] -3) <= pred.iloc[i]["max1_15d"] <= (test.iloc[i]["max1_15d"]+ 3)) == True):
    tol3_15d.append(1)
  else:
    tol3_15d.append(0)

  if (( (test.iloc[i]["max1_20d"] -1) <= pred.iloc[i]["max1_20d"] <= (test.iloc[i]["max1_20d"]+ 1)) == True):
    tol_20d.append(1)
  else:
    tol_20d.append(0)

  if (( (test.iloc[i]["max1_20d"] -2) <= pred.iloc[i]["max1_20d"] <= (test.iloc[i]["max1_20d"]+ 2)) == True):
    tol2_20d.append(1)
  else:
    tol2_20d.append(0)
  
  if (( (test.iloc[i]["max1_20d"] -3) <= pred.iloc[i]["max1_20d"] <= (test.iloc[i]["max1_20d"]+ 3)) == True):
    tol3_20d.append(1)
  else:
    tol3_20d.append(0)

print("5 Day Accuracy with 1 Day Tolerance for Max:",sum(tol_5d)/len(tol_5d))
print("10 Day Accuracy with 1 Day Tolerance for Max:",sum(tol_10d)/len(tol_10d))
print("10 Day Accuracy with 2 Day Tolerance for Max:",sum(tol2_10d)/len(tol2_10d))
print("15 Day Accuracy with 1 Day Tolerance for Max:",sum(tol_15d)/len(tol_15d))
print("15 Day Accuracy with 2 Day Tolerance for Max:",sum(tol2_15d)/len(tol2_15d))
print("15 Day Accuracy with 3 Day Tolerance for Max:",sum(tol3_15d)/len(tol3_15d))
print("20 Day Accuracy with 1 Day Tolerance for Max:",sum(tol_20d)/len(tol_20d))
print("20 Day Accuracy with 2 Day Tolerance for Max:",sum(tol2_20d)/len(tol2_20d))
print("20 Day Accuracy with 3 Day Tolerance for Max:",sum(tol3_20d)/len(tol3_20d))
```
