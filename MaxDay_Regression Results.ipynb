```python
import pandas as pd
import numpy as np

from sklearn.metrics import mean_squared_error
```

```python
# Load the data
ypred = pd.read_csv("3dayWindow_RF_Regression_ypred_Max.csv", header=None, names=['5d_Max', '10d_Max', '15d_Max', '20d_Max'])
ytest = pd.read_csv("3dayWindow_RF_Regression_ytest_Max.csv", header=None, names=['5d_Max', '10d_Max', '15d_Max', '20d_Max'])
```

```python
# Round ypred
ypred_round = ypred.round()

# No tolerance (Counter vars)
d5 = 0
d10 = 0
d15 = 0
d20 = 0

# Tolerances (Counter vars)
tol_5d = 0

tol_10d = 0
tol2_10d = 0

tol_15d = 0
tol2_15d = 0
tol3_15d = 0

tol_20d = 0
tol2_20d = 0
tol3_20d = 0

for i in range(len(ypred_round.index)):

  # No tolerance
  if (ypred_round.loc[i, "5d_Max"] == ytest.loc[i, "5d_Max"]):
    d5 += 1
  
  if (ypred_round.loc[i, "10d_Max"] == ytest.loc[i, "10d_Max"]):
    d10 += 1
  
  if (ypred_round.loc[i, "15d_Max"] == ytest.loc[i, "15d_Max"]):
    d15 += 1
  
  if (ypred_round.loc[i, "20d_Max"] == ytest.loc[i, "20d_Max"]):
    d20 += 1

  # 1 day tolerance
  if ((ypred_round.loc[i, "5d_Max"] - 1) <= ytest.loc[i, "5d_Max"] <= (ypred_round.loc[i, "5d_Max"] + 1)):
    tol_5d += 1
  
  if ((ypred_round.loc[i, "10d_Max"] - 1) <= ytest.loc[i, "10d_Max"] <= (ypred_round.loc[i, "10d_Max"] + 1)):
    tol_10d += 1
  
  if ((ypred_round.loc[i, "15d_Max"] - 1) <= ytest.loc[i, "15d_Max"] <= (ypred_round.loc[i, "15d_Max"] + 1)):
    tol_15d += 1
  
  if ((ypred_round.loc[i, "20d_Max"] - 1) <= ytest.loc[i, "20d_Max"] <= (ypred_round.loc[i, "20d_Max"] + 1)):
    tol_20d += 1

  # 2 day tolerance
  if ((ypred_round.loc[i, "10d_Max"] - 2) <= ytest.loc[i, "10d_Max"] <= (ypred_round.loc[i, "10d_Max"] + 2)):
    tol2_10d += 1
  
  if ((ypred_round.loc[i, "15d_Max"] - 2) <= ytest.loc[i, "15d_Max"] <= (ypred_round.loc[i, "15d_Max"] + 2)):
    tol2_15d += 1
  
  if ((ypred_round.loc[i, "20d_Max"] - 2) <= ytest.loc[i, "20d_Max"] <= (ypred_round.loc[i, "20d_Max"] + 2)):
    tol2_20d += 1

  # 3 day tolerance
  if ((ypred_round.loc[i, "15d_Max"] - 3) <= ytest.loc[i, "15d_Max"] <= (ypred_round.loc[i, "15d_Max"] + 3)):
    tol3_15d += 1
  
  if ((ypred_round.loc[i, "20d_Max"] - 3) <= ytest.loc[i, "15d_Max"] <= (ypred_round.loc[i, "20d_Max"] + 3)):
    tol3_20d += 1

# Accuracies
print("5 Day Accuracy:", d5/len(ypred_round.index))
print("10 Day Accuracy:", d10/len(ypred_round.index))
print("15 Day Accuracy:", d15/len(ypred_round.index))
print("20 Day Accuracy:", d20/len(ypred_round.index))
print("")
print("5 Day Accuracy with 1 Day Tolerance:", tol_5d/len(ypred_round.index))
print("10 Day Accuracy with 1 Day Tolerance:", tol_10d/len(ypred_round.index))
print("10 Day Accuracy with 2 Day Tolerance:", tol2_10d/len(ypred_round.index))
print("15 Day Accuracy with 1 Day Tolerance:", tol_15d/len(ypred_round.index))
print("15 Day Accuracy with 2 Day Tolerance:", tol2_15d/len(ypred_round.index))
print("15 Day Accuracy with 3 Day Tolerance:", tol3_15d/len(ypred_round.index))
print("20 Day Accuracy with 1 Day Tolerance:", tol_20d/len(ypred_round.index))
print("20 Day Accuracy with 2 Day Tolerance:", tol2_20d/len(ypred_round.index))
print("20 Day Accuracy with 3 Day Tolerance:", tol3_20d/len(ypred_round.index))
print("")
print("5 Day Highest Day Accuracy +/-", abs(ytest['5d_Max'] - ypred['5d_Max']).mean())
print("10 Day Highest Day Accuracy +/-", abs(ytest['10d_Max'] - ypred['10d_Max']).mean())
print("15 Day Highest Day Accuracy +/-", abs(ytest['15d_Max'] - ypred['15d_Max']).mean())
print("20 Day Highest Day Accuracy +/-", abs(ytest['20d_Max'] - ypred['20d_Max']).mean())
print("")
# RMSE
print("RMSE:", mean_squared_error(ytest, ypred, multioutput = 'raw_values', squared= False).tolist())
```
